# Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
iris = sns.load_dataset('iris')

iris.head()
```

# Dataset Understanding

```python
# Shape of dataset
print("Dataset Shape:")
print(iris.shape)

# Column names
print("\nColumns:") 
print(iris.columns)

# First 5 rows
print("\nFirst 5 Rows:")
print(iris.head())

# Statistical Summary
print("\nStatistical Summary:")
print(iris.describe())
```

# Data Cleaning & Preparation

```python
# Iris dataset is already clean (no missing values)
# But we still separate features and target

X = iris.drop('species', axis=1)
y = iris['species']
```

# Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=iris,
    x='petal_length',
    y='petal_width',
    hue='species',
    palette='Set2'
)
plt.title("Petal Length vs Petal Width by Species", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

**Histogram**

```python
plt.figure(figsize=(8, 6))
sns.histplot(iris['sepal_length'], kde=True, color='teal')
plt.title("Distribution Pattern of Sepal Lengths", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Sepal Length (cm)")
plt.show()
```

**Box Plot**

```python
plt.figure(figsize=(8,6))

sns.boxplot(data=iris)

plt.title("Feature Distribution & Outliers")
plt.show()
```
